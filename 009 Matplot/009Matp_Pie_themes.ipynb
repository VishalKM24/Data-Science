```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.pie([10,20,30], shadow = True)
```

```python
plt.pie([10, 20, 30], explode=[0.1, 0, 0], shadow = True, startangle=5)
```

- to change the angle use startangle = ....

```python
x = [10, 20, 30]
explode = [0.1, 0, 0]
labels=["Apple", "Google", "Facebook"]
colors = ["teal", "purple", "black"]
plt.pie(x, explode = explode,colors=colors,labels=labels, shadow = True)
plt.show()
```

```python
x = np.linspace(-40, 40, 100)
plt.plot(x, x**2)
plt.plot(x, x**3)
```

## circle

```python
theta = np.linspace(-np.pi, np.pi, 100)
rad = 10
```

```python
x = rad*np.cos(theta)
y = rad*np.sin(theta)
plt.plot(x, y)
```

- the above figure with scalling looks like circle but visual represent is like an oval

```python
plt.figure(figsize=(4, 4))
x = rad*np.cos(theta)
y = rad*np.sin(theta)
plt.plot(x, y)
plt.show()
```

- use figsize =( , ) in plt.figure() for scalling the page size

```python
theta = np.linspace(-np.pi, np.pi,1000)
rad = np.arange(0,100)
```

```python
plt.figure()
for i in rad:
    x = i * np.cos(theta)
    y = i * np.sin(theta)
    plt.plot(x, y)
plt.show()
```

```python
theta = np.linspace(0, 40*np.pi,1000)
rad = np.linspace(0, 100, 1000)
```

```python
plt.figure()
y = rad * np.sin(theta)
x = rad * np.cos(theta)
plt.plot(x, y)
plt.show()
```

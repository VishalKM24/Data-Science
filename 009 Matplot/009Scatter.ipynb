```python
mean = [10, 20]
covar = [[4,0],[0, 5]]
data = np.random.multivariate_normal(mean, covar, size=1000)
data.shape
```

```python
f1 = data[:,0]
f2 = data[:,1]
plt.scatter(f1, f2, marker="*")
```

```python
mean1 = [10, 20]
covar1 = [[4,0],[0, 5]]
data1 = np.random.multivariate_normal(mean1, covar1, size=100)
```

```python
mean2 = [30, 35]
covar2 = [[4,0],[0, 5]]
data2 = np.random.multivariate_normal(mean2, covar2, size=100)
```

```python
plt.scatter(data1[:, 0],data1[:,1], marker="*")
plt.scatter(data2[:, 0],data2[:,1], marker="^")
```

```python
data = np.vstack([data1, data2])
```

```python
plt.scatter(data[:,0], data[:,1], marker="*")
```

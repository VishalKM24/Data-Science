```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X , y = make_blobs(centers=3, random_state=42)
```

```python
X.shape, y.shape
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
model.classes_
```

```python
model.coef_
```

```python
model.intercept_
```

```python
colors = ["Red","Yellow","Blue","Black","Green"]
```

```python
y_colors = [colors[item] for item in y]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_colors)
```

```python
import seaborn as sn
```

```python
sn.scatterplot(X[:, 0], X[:, 1], hue=y_colors)
```

```python
list(zip(model.coef_, model.intercept_))
```

```python
for [m1, m2], c in zip(model.coef_, model.intercept_):
    
    x1 = np.linspace(-10, 10)
    x2 = -m1*x1/m2 - (c/m2)
    plt.plot(x1, x2)
```

```python
sn.scatterplot(X[:, 0], X[:, 1], hue=y_colors)

for [m1, m2], c in zip(model.coef_, model.intercept_):
    
    x1 = np.linspace(-10, 10)
    x2 = -m1*x1/m2 - (c/m2)
    plt.plot(x1, x2)
```

- to define colors according to index

```python
sn.scatterplot(X[:, 0], X[:, 1], hue=y_colors)

for index, [[m1, m2], c] in enumerate(zip(model.coef_, model.intercept_)):
    
    x1 = np.linspace(-10, 10)
    x2 = -m1*x1/m2 - (c/m2)
    plt.plot(x1, x2, color=colors[index])
    
plt.xlim(-10, 10)
plt.ylim(-10,10)
```

```python
model.score(X, y)
```

```python
model.predict_proba(X).argmax(axis=1)
```

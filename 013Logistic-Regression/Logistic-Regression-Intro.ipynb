## Logistic Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-500, 500, 100)
```

```python
y = 1 / (1 + np.exp(-x))
```

```python
plt.figure()
plt.plot(x, y)
plt.show()
```

```python
X = np.linspace(-2*np.pi, 2*np.pi)
```

```python
y1 = np.sin(X) + 2
```

```python
plt.plot(X, y1)
```

```python
plt.plot(X, np.log(y1))
```
